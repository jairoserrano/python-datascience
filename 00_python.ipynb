## Comprehensions y Generadores
Las comprehensions son una forma concisa de crear listas, conjuntos y diccionarios. Los generadores permiten crear iteradores personalizados de manera eficiente.

```python
# List comprehension
cuadrados = [x**2 for x in range(10)]

# Generador
def contador(maximo):
    i = 0
    while i < maximo:
        yield i
        i += 1

for num in contador(5):
    print(num)
```

## Decoradores y Context Managers
Los decoradores son funciones que modifican el comportamiento de otras funciones o mtodos. Los context managers permiten la gestin de recursos, como archivos, asegurando que se cierren correctamente.

```python
# Ejemplo de decorador
def decorador(funcion):
    def nueva_funcion(*args, **kwargs):
        print("Funcin decorada y funcionalidad adicional")
        return funcion(*args, **kwargs)
    return nueva_funcion

@decorador
def funcion_especial():
    print("Funcin original")

funcion_especial()

# Ejemplo de context manager
with open('archivo.txt', 'w') as archivo:
    archivo.write('Hola, mundo')
```
